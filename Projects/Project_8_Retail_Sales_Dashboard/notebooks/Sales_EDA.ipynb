# Retail Sales - Exploratory Data Analysis
Analyzing sales trends, regional performance, and category insights.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
```

## 1. Load Data

```
df = pd.read_csv('../data/retail_sales.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
df.head()
```

```
df.describe()
```

## 2. Monthly Revenue Trend

```
monthly = df.groupby(df['Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
monthly['Date'] = monthly['Date'].dt.to_timestamp()

plt.figure(figsize=(12, 5))
plt.plot(monthly['Date'], monthly['Revenue'], marker='o', color='#3498db', linewidth=2)
plt.fill_between(monthly['Date'], monthly['Revenue'], alpha=0.2, color='#3498db')
plt.title('Monthly Revenue Trend (2023)')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Revenue by Category

```
cat_rev = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']
plt.bar(cat_rev.index, cat_rev.values, color=colors, edgecolor='black', linewidth=0.5)
plt.title('Total Revenue by Category')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

## 4. Revenue by Region

```
reg_rev = df.groupby('Region')['Revenue'].sum()

plt.figure(figsize=(6, 6))
plt.pie(reg_rev, labels=reg_rev.index, autopct='%1.1f%%', colors=['#e74c3c', '#3498db', '#2ecc71', '#f39c12'])
plt.title('Revenue Distribution by Region')
plt.show()
```

## 5. Top Selling Products

```
top_prods = df.groupby('Product')['Revenue'].sum().sort_values(ascending=True)

plt.figure(figsize=(8, 5))
plt.barh(top_prods.index, top_prods.values, color='#9b59b6')
plt.title('Revenue by Product')
plt.xlabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

## 6. Heatmap: Region x Category

```
pivot = df.pivot_table(values='Revenue', index='Region', columns='Category', aggfunc='sum')

plt.figure(figsize=(8, 5))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Revenue Heatmap: Region vs Category')
plt.tight_layout()
plt.show()
```
