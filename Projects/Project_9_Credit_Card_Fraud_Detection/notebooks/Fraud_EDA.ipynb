# Credit Card Fraud - Exploratory Data Analysis
Analyzing class imbalance, feature distributions, and fraud patterns.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100
```

## 1. Load Data

```
df = pd.read_csv('../data/creditcard_sampled.csv')
print(f'Shape: {df.shape}')
df.head()
```

```
print('Class Distribution:')
print(df['Class'].value_counts())
print(f'\nFraud Rate: {df["Class"].mean():.2%}')
```

## 2. Class Imbalance Visualization

```
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.countplot(data=df, x='Class', ax=axes[0], palette=['#2ecc71', '#e74c3c'])
axes[0].set_title('Transaction Class Count')
axes[0].set_xticklabels(['Normal', 'Fraud'])

counts = df['Class'].value_counts()
axes[1].pie(counts, labels=['Normal', 'Fraud'], autopct='%1.2f%%', colors=['#2ecc71', '#e74c3c'])
axes[1].set_title('Class Proportion')

plt.tight_layout()
plt.show()
```

## 3. Amount Distribution

```
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(data=df[df['Class']==0], x='Amount', bins=50, ax=axes[0], color='#2ecc71', kde=True)
axes[0].set_title('Normal Transaction Amounts')

sns.histplot(data=df[df['Class']==1], x='Amount', bins=50, ax=axes[1], color='#e74c3c', kde=True)
axes[1].set_title('Fraudulent Transaction Amounts')

plt.tight_layout()
plt.show()
```

## 4. Feature Distributions

```
features = ['V1', 'V2', 'V3', 'V4', 'V5']

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for i, feat in enumerate(features):
    sns.boxplot(data=df, x='Class', y=feat, ax=axes[i], palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(feat)
    axes[i].set_xticklabels(['Normal', 'Fraud'])

plt.tight_layout()
plt.show()
```

## 5. Correlation Heatmap

```
plt.figure(figsize=(8, 6))
corr = df.drop('TransactionID', axis=1).corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## 6. Scatter Plot: V1 vs Amount

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='V1', y='Amount', hue='Class', alpha=0.5, palette={0:'#2ecc71', 1:'#e74c3c'})
plt.title('V1 vs Amount (Colored by Class)')
plt.legend(['Normal', 'Fraud'])
plt.tight_layout()
plt.show()
```
