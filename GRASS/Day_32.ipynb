# Unsupervised learning
Types of unsupervised :
1. Clustering concepts :\
   -> clustering means grouping similiar data points together.\
   -> sorting object
   1. K-Means Clustering :\
      -> most popular clustering algorithm\
      -> It divides data into K clusters\
      -> k --> no of clusters\
      note:- Elbow method -> helps to find the best value of k. 
   
   

```python
# clustering
import pandas as pd 
from sklearn.cluster import KMeans

data = {
    "age": [22,25,24,45,50,48,20,47],
    "income": [22000,25000,24000,45000,50000,48000,20000,47000]
}

df = pd.DataFrame(data)
print(df)

# input| feature 
x = df[['age','income']]

# model
model = KMeans(n_clusters=2,random_state=42)
model.fit(x)

# clusters
clusters = model.labels_
print(clusters)
df['clusters'] = clusters
print()
```

```python
# example
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = {
    "age": [22,25,24,45,50,48,20,47],
    "income": [22000,25000,24000,45000,50000,48000,20000,47000]
}

df = pd.DataFrame(data)
print(df)

# input| feature 
x = df[['age','income']]

# model
model = KMeans(n_clusters=2,random_state=42)
model.fit(x)

# clusters
df['clusters'] = model.labels_

plt.scatter(df['age'],df['income'],c=df['clusters'])
plt.xlabel('age')
plt.ylabel('income')
plt.title('Group the Data')
plt.show()
```

```python
# example 
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()

X = iris.data

# KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

clusters = kmeans.fit_predict(X)

# Plot
plt.scatter(
    X[:, 0],
    X[:, 1],
    c=clusters,
    cmap="viridis"
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K-Means on Iris Dataset")

plt.show()
```

# Deep Learning
Deep learning is a subset of ML that uses ANN( Artificial Neural Network) with multiple hidden layers to learn pattern from data.

DL architecture:\
Input layer---->hl1-->hl2-->hl3----->output layer
receives input | process information| final prediction
data.          | learn pattern      |

-------------------------------------------------------
types of DL:-
1. ANN :- artificial neural network
2. CNN :- Covolutional neural network
3. RNN :- recurrent neural network
-------------------------------------------------
1. ANN :-
   -> basic neural network.\
   -> It is inspired by the human brain.\
   -> Every neuron receives input.\
   -> Perform calculation/process/pattern.\
   -> And passes the output layer.\
    Input data----> calculation pattern -----> Output data
       Layer            H.L. process             Layer
\
   Uses of ANN:-
   1. House price prediction
   2. Sales Prediction
   3. Loan Approval
   4. Stock
------------------------------------------------------------------------
2. CNN :- Covolutional neural network\
   -> CNN is specially designed for Image process\
   -> It automatically detects - edges,shapes,object,complete image
\
   Uses of CNN:-
   1. Face recognition
   2. medical imaging
   3. Self driving

----------------------------------------------------------------------
3. RNN :- recurrent neural network\
   -> RNN is used for sequential data.\
   -> it resembles previous information.\
   -> ex. - sentences, music, speech, stock prices, house prices\
   -> RNN has memory.\

   Uses of RNN:-
   1. Chatbots
   2. Speech Recognition
   3. Language Translation
   4. Time Series Forecasting
----------------------------------------------------------------------

4. LSTM :- Long short term memory
   -> LSTM is an improved version of RNN.
   -> It remembers information for a long time.

-----------------------------------------------------------------------
5. GRU :- Gated Recurrent Unit
   -> Improved version of RNN.
